```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Crime = pd.read_csv("crime_data.csv")
Crime
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(Crime.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(Crime.iloc[:,1:]))
data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel ('INDEX')
plt.ylabel ('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel ('INDEX')
plt.ylabel ('Distance')
sch.dendrogram(z)
plt.show()
```

```python
#Considering Dendrogram made of complete linkage as it shows more appropriate results - No of Clusters: 6
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=6, linkage='complete',affinity = "euclidean").fit(df_norm) 
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
Crime['clust']=cluster_labels # creating a  new column for clusters and assigning it to new column 
Crime
```

```python
Crime.iloc[:,1:].groupby(Crime.clust).mean()
```

```python
data = Crime[(Crime.clust==1)]
data
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
# Kmeans Clustering elbow curve also shows no of clusters as 6
```

```python
clf = KMeans(n_clusters=6)
y_kmeans = clf.fit_predict(df_norm)
```
